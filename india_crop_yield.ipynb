# Data Preprocessing

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
df = df.dropna()
```

```python
df['production'] = df['production'].replace('=', np.nan)
df.dropna(subset=['production'], inplace=True)
```

```python
df.head()
```

```python
df2 = df[['state_names','crop_year','crop_names','temperature','precipitation','humidity','N','P','K','production']]
df2.head()
```

```python
df2.dtypes
```

```python
df2['production'] = df2['production'].astype(float)
df2.dtypes
```

### Encoding Categorical Variables



```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode_df2 = pd.get_dummies(df2, columns=['state_names',"crop_names"], prefix = ['State',"Crop"])
features=encode_df2.loc[:, encode_df2.columns != 'production']
label=df2['production']
features.head()
```

```python
features = features.drop(['crop_year'], axis=1)
```

```python
features.info()
```

```python
features.head()
```

```python
features = features.replace({True: 1, False: 0})

# Verify the conversion
features.head()
```


### Scaling Features



```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
features=scaler.fit_transform(features) 
```

```python
features
```

## Training Data: 






```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_labels, test_labels = train_test_split(features, label, test_size=0.2, random_state=42)
```

# Compare 4 models with the training data 
 
- Random Forest Regressor
- Decision Tree Regressor
- SVM 
- Gradient Boosting Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
model.fit(train_data, train_labels)

# Predict on the test data
predictions = model.predict(test_data)

# Evaluate the model
rf_mse = mean_squared_error(test_labels, predictions)
rf_r2 = r2_score(test_labels, predictions)

print(f"Random Forest Regressor - Mean Squared Error: {rf_mse}")
print(f"Random FOrest Regressor - R^2 Score: {rf_r2}")

# Example: Predicting for new data (optional)
# new_data = [[...]  # Example new data
# new_prediction = model.predict(new_data)
# print(f"Predicted crop yield: {new_prediction}")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42)

# Train the model on the training data
dt_model.fit(train_data, train_labels)

# Predict on the test data
dt_predictions = dt_model.predict(test_data)

# Evaluate the model
dt_mse = mean_squared_error(test_labels, dt_predictions)
dt_r2 = r2_score(test_labels, dt_predictions)

print(f"Decision Tree Regressor - Mean Squared Error: {dt_mse}")
print(f"Decision Tree Regressor - R^2 Score: {dt_r2}")
```

```python
from sklearn.svm import SVR

# Initialize the SVM Regressor
svm_model = SVR()

# Train the model on the training data
svm_model.fit(train_data, train_labels)

# Predict on the test data
svm_predictions = svm_model.predict(test_data)

# Evaluate the model
svm_mse = mean_squared_error(test_labels, svm_predictions)
svm_r2 = r2_score(test_labels, svm_predictions)

print(f"SVM Regressor - Mean Squared Error: {svm_mse}")
print(f"SVM Regressor - R^2 Score: {svm_r2}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialize the Gradient Boosting Regressor
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
gb_model.fit(train_data, train_labels)

# Predict on the test data
gb_predictions = gb_model.predict(test_data)

# Evaluate the model
gb_mse = mean_squared_error(test_labels, gb_predictions)
gb_r2 = r2_score(test_labels, gb_predictions)

print(f"Gradient Boosting Regressor - Mean Squared Error: {gb_mse}")
print(f"Gradient Boosting Regressor - R^2 Score: {gb_r2}")
```
